```python
#baseline dummy
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

- `Example UseCASE`
    `from sklearn.dummy import DummyClassifier`
    `import numpy as np`

#### Example training data
    `X_train = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])`
    `y_train = np.array([0, 1, 0, 1])`
    `dummy = DummyClassifier(strategy="most_frequent")`  # Initializes with a strategy
    `dummy.fit(X_train, y_train)`  # Trains the classifier on the training data


### What Happens During fit()?
    -No Learning of Patterns: The fit() method doesn't actually learn complex patterns or relationships between X_train and y_train. Instead, it computes the information needed to apply the selected strategy.

    -For example, with strategy="most_frequent", the classifier will learn that the most frequent class in y_train is 0, and it will always predict 0 for any input during inference.
    
`Predicting with the Trained DummyClassifier:`
    -Once the classifier is fitted, you can use it to make predictions:

- y_pred = dummy.predict(X_train)
  - print(y_pred)

- Output for strategy="most_frequent":
    array([0, 0, 0, 0])

- Why Use a DummyClassifier?
    Baseline Performance: It provides a baseline to compare your more sophisticated models against. If your complex model performs worse than a DummyClassifier, it suggests that the model might not be capturing meaningful patterns.
- Debugging: It helps ensure that your data pipeline and evaluation metrics are working as expected.

```python
#Baseline Dummy
from sklearn.dummy import DummyClassifier
dummy_clf = DummyClassifier()
dummy_clf.fit(X_train,y_train)
```

When testing accuracy is higher than training accuracy, it’s an unusual situation but can occur under certain conditions. Here are a few potential reasons for this:

- `Data Leakage:` There might be data leakage where the testing data inadvertently includes information from the training data. This would lead to an inflated testing accuracy.

- `Small Sample Size:` If the training set is very small, the model might not have learned enough to perform well on it, but it could perform surprisingly well on the test set if the test set happens to be more aligned with the model's learned patterns.

- `Random Fluctuations:` Sometimes, due to randomness in data splitting or the inherent variability in small datasets, you might see fluctuations where the test accuracy is temporarily higher than the training accuracy.

- `Overfitting Not Present:' In some cases, the model might not be overfitting, and the test data might be less noisy or more representative than the training data.

- `Evaluation Metrics Issue:` Make sure you’re correctly splitting the data and using the right metrics. Sometimes mistakes in preprocessing or evaluation can lead to misleading results.
