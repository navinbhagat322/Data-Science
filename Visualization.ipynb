```python
fig,axs=plt.subplots(1,2)  #( i wanted to create two blocks from a bigger block so row=1, column = 2)
axs[0].plot(a,c)  # axis[0] means plot 1st subplot as python read indexes from 0
axs[1].plot(a,b)
plt.show()
```

###### axs: This is an array of Axes objects. In this case, plt.subplots(1, 2) creates a figure with 1 row and 2 columns of subplots, resulting in an array of 2 Axes objects. These subplots can be individually accessed using indexing, e.g., axs[0] for the first subplot and axs[1] for the second subplot.

###### fig: This is a Figure object. In Matplotlib, a figure is the entire window or page that contains all the plot elements. It can contain multiple Axes (subplots), titles, labels, and other plot components.

```python
fig,axs = plt.subplots(1,2)
axs[0].plot(a,c,marker='o',linestyle='-',color='b')
axs[0].set_title("A,C performance")
axs[1].plot(a,b,marker='*',linestyle='-.',color='y')
axs[1].set_title("A,B perfomance")
plt.show()
```

### another way of same thing 

```python
plt.subplot(221)  # 2 rows 2 column 1st graph
plt.plot(a,c)
plt.title("graph 1")

plt.subplot(222)   # 2 rows 2 column 2nd graph
plt.scatter(a,c)
plt.title("graph2")

plt.subplot(223)
plt.scatter(b,c)
plt.title("graph3")

plt.subplot(224)
plt.plot(b,c,marker='*',linestyle='--',color='y')
plt.title("graph 4")

plt.tight_layout()
plt.show()
```

# BAR Plot (For categorical data it is used)

```python
dict1={'lenovo':10, 'mac':5 ,'Dell':8}
x=dict1.keys()
y=dict1.values()
plt.bar(x,y)
plt.show()
```

```python
dict1={'lenovo':10, 'mac':5 ,'Dell':8}
x=list(dict1.keys())
y=list(dict1.values())
plt.bar(x,y,color=['yellow','green','red'])
plt.show()
```

## Pie Chart ( uses for categorical data but it's readibility decreases as category>10 and in that case we will prefer bar plot )

```python
plt.pie(y,labels=x)
x=list(dict1.keys())
y=list(dict1.values())
plt.show()
```
